### importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### importing and manipulating train data

```python
train_data = pd.read_csv('data/train.csv')
train_data.head()
```

```python
train_data.info()
```

```python
sex = train_data.iloc[:,4:5].values.reshape(-1)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
sex = le.fit_transform(sex)
sex = pd.DataFrame(sex)
sex
```

```python
age = train_data.iloc[:,5:6]
age
```

```python
age.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(age)
age = imputer.transform(age)
age = pd.DataFrame(age)
age
```

```python
age.isna().sum()
```

```python
p_class = train_data.iloc[:,2:3]
```

```python
s1 = pd.concat([p_class, age], axis=1)
x_train = pd.concat([s1, sex], axis=1).values
x_train
```

```python
y_train = train_data.iloc[:,1:2].values.reshape(-1)
```

### importing and manipulating test data

```python
test_data = pd.read_csv('data/test.csv')
test_data.head()
```

```python
test_data.info()
```

```python
sex_test = test_data.iloc[:,3:4].values.reshape(-1)
from sklearn.preprocessing import LabelEncoder
le2 = LabelEncoder()
sex_test = le.fit_transform(sex_test)
sex_test = pd.DataFrame(sex_test)
sex_test
```

```python
age_test = test_data.iloc[:,4:5].values
from sklearn.impute import SimpleImputer
imputer2 = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer2 = imputer2.fit(age_test)
age_test = imputer2.transform(age_test)
age_test = pd.DataFrame(age_test)
age.isna().sum()
age
```

```python
p_class_test = test_data.iloc[:,1:2]
```

```python
s2 = pd.concat([p_class_test, age_test], axis=1)
x_test = pd.concat([s2, sex_test], axis=1).values
x_test
```

### scaling datas

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
x_train_scaled = ss.fit_transform(x_train)
x_test_scaled = ss.transform(x_test)
```

### logistic regression model

```python
from sklearn.linear_model import LogisticRegression

logr = LogisticRegression()
logr.fit(x_train_scaled, y_train)
y_pred_logr = logr.predict(x_test_scaled)
y_pred_logr
```

```python
predicted = pd.DataFrame(y_pred_logr)
predicted
```

```python
pass_id = test_data.iloc[:,0]
pass_id
```

```python
final = pd.concat([pass_id, predicted], axis=1)
final.columns = ['PassengerId', 'Survived']
final
```

```python
final.to_csv('submission.csv', index=False)
```
